```python
mydb = pymysql.connect(
    host="127.0.0.1", 
    user="root", 
    password="MySQL123@", 
    database="zomatto"
)
```

```python
mycursor = mydb.cursor()
```

```python
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS customer_table (
        customer_id INT PRIMARY KEY AUTO_INCREMENT,
        name VARCHAR(100),
        email VARCHAR(50),
        phone VARCHAR(30),
        location VARCHAR(250),
        is_premium BOOLEAN,
        preferred_cuisine VARCHAR(50),
        total_orders INT,
        average_rating FLOAT
    )
""")
mydb.commit()
```

```python
customer_df = data_instance.customer_table()
```

```python
insert_query = """
    INSERT INTO customer_table(name, email, phone, location, 
     is_premium, preferred_cuisine, total_orders, average_rating)
    VALUES(%s, %s, %s, %s, %s, %s, %s, %s)
"""
for index, row in customer_df.iterrows():
    mycursor.execute(insert_query, (
        row['name'], row['email'], row['phone'], row['location'],
         row['is_premium'], row['preferred_cuisine'], row['total_orders'], row['average_rating']
    ))
mydb.commit()
```

```python
mycursor.execute("SELECT * FROM customer_table")
```

```python
myresults = mycursor.fetchall()
for x in myresults:
    print(x)
```

```python
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS restaurant_table(
        restaurant_id INT PRIMARY KEY AUTO_INCREMENT,
        name VARCHAR(100),
        cuisine_type VARCHAR(50),
        location VARCHAR(250),
        owner_name VARCHAR(100),
        average_delivery_time INT,
        rating INT,
        total_orders INT,
        is_active BOOLEAN
    )
""")
mydb.commit()
```

```python
restaurant_df = data_instance.restaurant_table()
```

```python
insert_query = """
    INSERT INTO restaurant_table(name, cuisine_type, location, owner_name,
    average_delivery_time, rating, total_orders, is_active)
    VALUES(%s, %s, %s, %s, %s, %s, %s, %s)
"""
for index, row in restaurant_df.iterrows():
    mycursor.execute(insert_query, (
        row['name'], row['cuisine_type'], row['location'], row['owner_name'],
        row['average_delivery_time'], row['rating'], row['total_orders'], row['is_active']
    ))
mydb.commit()
```

```python
mycursor.execute("SELECT * FROM restaurant_table")
```

```python
myresults = mycursor.fetchall()
for x in myresults:
    print(x)
```

```python
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS order_table(
        order_id INT PRIMARY KEY AUTO_INCREMENT,
        customer_id INT,
        restaurant_id INT,
        order_date DATE,
        delivery_time INT,
        status VARCHAR(50),
        total_amount INT,
        payment_mode VARCHAR(50),
        discount_applied VARCHAR(20),
        feedback_rating INT,
        FOREIGN KEY(customer_id) REFERENCES customer_table(customer_id),
        FOREIGN KEY(restaurant_id) REFERENCES restaurant_table(restaurant_id)
    )
""")
mydb.commit()
```

```python
order_df = data_instance.order_table()
```

```python
insert_query = """
    INSERT INTO order_table(customer_id, restaurant_id, order_date, delivery_time,
    status, total_amount, payment_mode, discount_applied, feedback_rating)
    VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
for index, row in order_df.iterrows():
    mycursor.execute(insert_query, ( 
        row['customer_id'], row['restaurant_id'], row['order_date'], row['delivery_time'],
        row['status'], row['total_amount'], row['payment_mode'], row['discount_applied'], row['feedback_rating']
    ))

mydb.commit()
```

```python
mycursor.execute("SELECT * FROM order_table")
```

```python
myresults = mycursor.fetchall()
for x in myresults:
    print(x)
```

```python
mycursor.execute("""
    CREATE TABLE IF NOT EXISTS delivery_table(
        delivery_id INT PRIMARY KEY AUTO_INCREMENT,
        order_id INT,
        delivery_person_id VARCHAR(20),
        delivery_status VARCHAR(50),
        distance INT,
        delivery_time INT,
        estimated_time INT,
        delivery_fee INT,
        vehicle_type VARCHAR(20),
        FOREIGN KEY(order_id) REFERENCES order_table(order_id)
    )
""")
mydb.commit()
```

```python
delivery_df = data_instance.delivery_table()
```

```python
insert_query = """
    INSERT INTO delivery_table(order_id, delivery_person_id, delivery_status, distance,
    delivery_time, estimated_time, delivery_fee, vehicle_type)
    VALUES(%s, %s, %s, %s, %s, %s, %s, %s)
"""
for index, row in delivery_df.iterrows():
    mycursor.execute(insert_query, (
        row['order_id'], row['delivery_person_id'], row['delivery_status'], row['distance'],
        row['delivery_time'], row['estimated_time'], row['delivery_fee'], row['vehicle_type']
    ))
mydb.commit()
```
